# <p style="text-align: center;">Data Science interests classification</p>

A survey was conducted to gauge an audience interest in different data science topics, namely:

    Big Data (Spark / Hadoop)
    Data Analysis / Statistics
    Data Journalism
    Data Visualization
    Deep Learning
    Machine Learning

The participants had three options for each topic: Very Interested, Somewhat interested, and Not interested. 2,233 respondents completed the survey.

```python
#Importing primary modules
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#First of all loadig the csv file into the dataframe
#link-location => https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/Topic_Survey_Assignment.csv
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/Topic_Survey_Assignment.csv', index_col=0)
print("dataset successfully loaded")
```

```python
df.head()
```

Visualizing the different sets of people in percentages in graph

```python
#Sort the dataframe in descending order of Very interested.
df = df.sort_values(by =['Very interested'], ascending = False)

#Convert the numbers into percentages of the total number of respondents.
total_people = 2233

df_percent = (100. * df / total_people).round(2)
ax = df_percent.plot(kind='bar', figsize = (20,8), width = 0.8, color = ('#5cb85c', '#5bc0de','#d9534f'), fontsize=14)
plt.legend(fontsize = 14)
plt.title("Percentage of Respondents' Interest in Data Science Areas", fontsize = 16)
#removing y-axis label
plt.yticks([])

#removing border
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)

#adding percent signs
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')
```
